```python
from sklearn.cluster import KMeans
import numpy as np 
```

```python
a=np.array([[2,10],[2,5],[8,4],[5,8],[7,5],[6,4],[1,2],[4,9]])
Kmeans = KMeans(n_clusters=3, random_state=0).fit(a)
```

```python
Kmeans.labels_
```

```python
Kmeans.predict([[2,7]])
```

```python
Kmeans.cluster_centers_
```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Mall_Customers.csv")
data
```

```python
scaler = MinMaxScaler()
scale = scaler.fit_transform(data[['Annual Income (k$)','Spending Score (1-100)']])
data_scale = pd.DataFrame(scale, columns = ['Annual Income (k$)','Spending Score (1-100)']);
data_scale
```

```python
km=KMeans(n_clusters=2)
y_predicted = km.fit_predict(data_scale[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

```python
km.cluster_centers_
```

```python
data['Clusters'] = km.labels_
sns.scatterplot(x="Spending Score (1-100)", y="Annual Income (k$)",hue = 'Clusters',  data=data,palette='viridis')
```

```python
d=pd.read_csv("diabetes.csv")
d
```

```python
d.drop('Outcome',axis=1)
```

```python
scaler = MinMaxScaler()
scale = scaler.fit_transform(d[['Glucose','BloodPressure','DiabetesPedigreeFunction']])
data_scale = pd.DataFrame(scale, columns = ['Glucose','BloodPressure','DiabetesPedigreeFunction']);
data_scale
```
